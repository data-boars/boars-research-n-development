```python
from PIL import Image
import numpy as np
import cv2
from moviepy import editor as me
import moviepy.video.fx.all as vfx
from src.algorithms.cv2_detecting import ContourDetector, ContourFilter, draw_bboxes
```

```python
target_res = (None, 480)
video = me.VideoFileClip('./data/quadracopter_sample/videos/dzik25fps.mp4',
                         target_resolution=target_res, audio=False)
video = video.cutout('00:00:50.0', '00:01:15.0')
video = video.cutout('00:01:31.0', '00:02:17.0')

video = video.set_duration(15)

video = video.fx(vfx.blackwhite)
```

```python
me.ipython_display(clip=video)
```

```python
contour_filter = ContourFilter(hard_threshold=102,
                               soft_threshold=98,
                               shape_ratio_threshold=1.6,
                               common_area_threshold=15)
detector = ContourDetector()
video_analysed = video.fl_image(lambda frame: draw_bboxes(frame, 
                                                          contour_filter.filter_contours(frame, 
                                                                                     detector.detect_animal_contours(frame))))
me.ipython_display(clip=video_analysed)
```
